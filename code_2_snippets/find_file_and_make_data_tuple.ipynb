```python
import os
import unicodedata
import shutil
dst_for_tuple = "../../../data/Dataset/"
```

```python
cnt =0
notfound_list = []
for file in os.listdir("../../../data/csv/csv_seper_res/"):
    flag =0
    for (path, dir, imgfile) in os.walk("../../../gen4test/image/") :
        for filename in imgfile:
            ext = os.path.splitext(filename)[-1]
            if ext == ".jpeg":
                uni1 = unicodedata.normalize('NFD', file[0:-8])
                uni2 = unicodedata.normalize('NFD', filename[0:-5])
                if(uni1 == uni2 and (file[0:-8] != '.' and file[0:-8] != ' ' and file[0:-8] != '')):
                    print(file[0:-8],"<- file[0:-8] filename-> ", filename[0:-5])
                    src_csv = os.path.join("../../../data/csv/csv_seper_res/",file)
                    src_img = os.path.join(path, filename)
                    mkdir_new = os.path.join(dst_for_tuple,file[0:-8])
                    try : 
                        if not os.path.exists(mkdir_new):
                            os.makedirs(mkdir_new)
                    except OSError:
                        print('Error: Creating directory. ' + mkdir_new)
                    src_csv = os.path.realpath(src_csv)
                    src_img = os.path.realpath(src_img)
                    shutil.copy(src_img, mkdir_new)
                    shutil.copy(src_csv, mkdir_new)
                    cnt = cnt+1
                    flag=1
                    break
    #if flag ==0: print("shit")     
    #print(file_path)
    if flag==0:
        print(file)    
        notfound_list.append(file)
    
    #file_path = file_path.replace("\\", "/")


```
